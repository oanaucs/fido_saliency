```python
from deeplift_compare import run
from deeplift_compare import parse_args
```

```python
args = parse_args()
print(args)
```

### Test Ising Soft Penalty with different scale of l1 and l2 with grid search

```python
args.top_n = 3
args.verbose = 0
```

```python
# Test L1 penalty range from 1E-5, 1E-3, 0.01, 0.1, 1., 5.
args.l2_reg_coef = 0.

l1_ranges = [0., 1E-3, 0.01, 0.1, 1., 5.]
for l1_reg in l1_ranges:
    args.l1_reg_coef = l1_reg
    diff = run(args)
    print 'l1 reg %f, total odds change: %f' % (l1_reg, sum(diff))
```

```python
# Test L2 penalty range from 1E-5, 1E-3, 0.01, 0.1, 1., 5.
args.l1_reg_coef = 0.

l2_ranges = [0., 1E-5, 1E-3, 0.01, 0.1, 1., 5.]
for l2_reg in l2_ranges:
    args.l2_reg_coef = l2_reg
    diff = run(args)
    print 'l2 reg %f, total odds change: %f' % (l2_reg, sum(diff))
```
